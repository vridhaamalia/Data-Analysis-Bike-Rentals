## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Group by workingday dan holiday untuk menghitung jumlah total penyewaan
day_type_agg = data.groupby(by=["workingday", "holiday"]).agg({"cnt": "sum"}).reset_index()

# Buat kolom 'day_type' untuk menggabungkan informasi workingday dan holiday
day_type_agg['day_type'] = day_type_agg.apply(lambda row: 'Holiday' if row['holiday'] == 1 else 'Working Day' if row['workingday'] == 1 else 'Weekend', axis=1)

# Plot barplot untuk frekuensi penyewaan sepeda berdasarkan tipe hari
plt.figure(figsize=(8, 6))
sns.barplot(x='day_type', y='cnt', data=day_type_agg, palette='Set2')
plt.title('Total Bike Rentals by Day Type (Working Day, Weekend, Holiday)')
plt.xlabel('Day Type')
plt.ylabel('Total Rentals')
plt.show()
```

```python
# Buat kolom 'day_type' untuk menggabungkan informasi workingday dan holiday
day_type_agg['day_type'] = day_type_agg.apply(lambda row: 'Holiday' if row['holiday'] == 1 else 'Working Day' if row['workingday'] == 1 else 'Weekend', axis=1)

# Plot line chart untuk frekuensi penyewaan sepeda berdasarkan tipe hari
plt.figure(figsize=(8, 6))
sns.lineplot(x='day_type', y='cnt', data=day_type_agg, marker='o', linestyle='-', color='blue')
plt.title('Total Bike Rentals by Day Type (Weekend, Holiday, Working Day)')
plt.xlabel('Day Type')
plt.ylabel('Total Rentals')
plt.show()
```

```python
# Buat kolom 'day_type' untuk menggabungkan informasi workingday dan holiday
day_type_agg['day_type'] = day_type_agg.apply(lambda row: 'Holiday' if row['holiday'] == 1 else 'Working Day' if row['workingday'] == 1 else 'Weekend', axis=1)

# Mengelompokkan lagi berdasarkan 'day_type' untuk mendapatkan total per tipe hari
day_type_total = day_type_agg.groupby('day_type')['cnt'].sum().reset_index()

# Membuat Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(day_type_total['cnt'], labels=day_type_total['day_type'], autopct='%1.1f%%', startangle=90, colors=['#66b3ff', '#99ff99', '#ffcc99'])
plt.title('Proportion of Bike Rentals by Day Type (Working Day, Weekend, Holiday)')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()
```

### Pertanyaan 2:

```python
plt.figure(figsize=(12, 6))

# Lineplot untuk melihat pola penyewaan berdasarkan kondisi cuaca
sns.lineplot(x='weathersit', y='cnt', data=data, marker='o', ci=None)
plt.title('Trend of Bike Rentals by Weather Situation')
plt.xlabel('Weather Situation (1: Clear, 2: Mist, 3: Light Rain/Snow, 4: Heavy Rain/Snow)')
plt.ylabel('Total Rentals')
plt.xticks([1, 2, 3, 4], ['Clear', 'Mist', 'Light Rain/Snow', 'Heavy Rain/Snow'])

plt.tight_layout()
plt.show()
```

```python
# Mengubah angka kategori cuaca menjadi label yang lebih mudah dibaca
weather_agg['weather_label'] = weather_agg['weathersit'].map({1: 'Clear/Few Clouds',2: 'Mist/Cloudy', 3: 'Light Rain/Snow',4: 'Heavy Rain/Snow'})

# Plot Barplot untuk jumlah total penyewaan berdasarkan kondisi cuaca
plt.figure(figsize=(8, 6))
sns.barplot(x='weather_label', y='cnt', data=weather_agg, palette='Blues_d')
plt.title('Total Bike Rentals by Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Total Rentals')
plt.xticks(rotation=45)
plt.show()
```

**Insight:**
- From bar chart, line plot and pie chart we know that Working Days have the highest number of rentals compared to weekends and holidays, we also know that Clear Weather leads to the most bike rentals, while poor weather conditions like Heavy Rain or Snow drastically reduce the number of rentals.
